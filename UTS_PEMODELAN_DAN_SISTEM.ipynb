```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks transisi (contoh probabilitas transisi antar status)
transition_matrix = np.array([
    [0.6, 0.3, 0.1],  # Probabilitas transisi dari Awal
    [0.2, 0.5, 0.3],  # Probabilitas transisi dari Menengah
    [0.1, 0.2, 0.7],  # Probabilitas transisi dari Maju
])

# Status awal (misalnya startup saat ini dalam keadaan "Awal")
initial_state = np.array([1, 0, 0])  # 100% berada di status "Awal"

# Fungsi untuk menghitung prediksi status di masa depan dan menampilkan grafik
def predict_growth_with_matrix_and_plot(transition_matrix, initial_state, years):
    state = initial_state
    # Menyimpan hasil prediksi untuk setiap tahun
    states_over_time = []

    # Lakukan prediksi selama beberapa tahun
    for year in range(years):
        state = np.dot(state, transition_matrix)
        states_over_time.append(state)

    # Konversi hasil prediksi menjadi array numpy untuk kemudahan plotting
    states_over_time = np.array(states_over_time)

    # Menampilkan hasil matriks prediksi (output matriks)
    print(f"Prediksi Pertumbuhan Startup EdTech (Matriks Probabilitas):")
    for i, state in enumerate(states_over_time):
        print(f"Tahun {i+1}: {state}")

    # Menampilkan grafik
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, years+1), states_over_time[:, 0], label='Awal', marker='o')
    plt.plot(range(1, years+1), states_over_time[:, 1], label='Menengah', marker='o')
    plt.plot(range(1, years+1), states_over_time[:, 2], label='Maju', marker='o')

    plt.title('Prediksi Pertumbuhan Startup EdTech di Indonesia (Markov Chain)', fontsize=14)
    plt.xlabel('Tahun', fontsize=12)
    plt.ylabel('Probabilitas', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Lakukan prediksi untuk 5 tahun ke depan dan tampilkan grafik serta matriks
predict_growth_with_matrix_and_plot(transition_matrix, initial_state, 5)
```
